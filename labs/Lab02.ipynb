## 1. Leer una imagen

```python
I = plt.imread('quetzal.png')
I.shape
```

```python
type(I)
```

El shape es un vector de la forma (h,w,c), con

* h = número de filas (altura)
* w = número de columnas (ancho)
* c = número de canales de información (c=3, en el caso de imágenes RGB)

En este caso, una imagen RGB consiste de tres canales:

* R = canal rojo
* G = canal verde
* B = canal azul

## 2. Canales de color

Descomponemos la imagen en sus tres canales de color.

```python
IR = I[:,:,0]
IG = I[:,:,1]
IB = I[:,:,2]

print(IR.shape, IG.shape, IB.shape)
```

```python
plt.figure(figsize=(17, 5.5))
plt.subplot(1,3,1)
plt.imshow(IR)
plt.xlabel('Canal R')

plt.subplot(1,3,2)
plt.imshow(IG)
plt.xlabel('Canal G')

plt.subplot(1,3,3)
plt.imshow(IB)
plt.xlabel('Canal B')
plt.show()
```

Las imágenes se ven a color porque matplotlib por default les asigna un mapa de colores ('jet').

Vamos a mostrar cada imagen en escala de grises.

```python
plt.figure(figsize=(17, 5.5))
plt.subplot(1,3,1)
plt.imshow(IR, cmap='gray')
plt.xlabel('Canal R')
plt.subplot(1,3,2)
plt.imshow(IG, cmap='gray')
plt.xlabel('Canal G')
plt.subplot(1,3,3)
plt.imshow(IB, cmap='gray')
plt.xlabel('Canal B')
plt.show()
```

Se pueden aplicar otros mapas de color. Por ejemplo

```python
plt.figure(figsize=(17, 5.5))
plt.subplot(1,3,1)
plt.imshow(IR, cmap=plt.cm.Reds_r)
plt.xlabel('Canal R')
plt.subplot(1,3,2)
plt.imshow(IG, cmap=plt.cm.Greens_r)
plt.xlabel('Canal G')
plt.subplot(1,3,3)
plt.imshow(IB, cmap=plt.cm.Blues_r)
plt.xlabel('Canal B')
plt.show()
```

## 3. Convertir una imagen a escala de grises

```python
#gray = 0.333R + 0.333G + 0.333B

Igray = I.mean(axis=2)
Igray.shape
```

```python
plt.figure(figsize=(5,5.5))
plt.imshow(Igray, cmap='gray')
plt.show()
```

No es la mejor forma, típicamente se pondera cada canal da forma diferente, con el objetivo de preservar ciertas propiedades de luminosidad.

```python
# otros ejemplos usuales de convertir a grises
#Y = 0.299R + 0.587G + 0.114B
#Y = 0.2126R + 0.7152G + 0.0722B
#Y = 0.2627R + 0.6780G + 0.0593B

Igray2 = rgb2gray(I)
Igray2.shape
```

```python
plt.figure(figsize=(11,5.5))
plt.subplot(1,2,1)
plt.imshow(Igray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(Igray2, cmap='gray')
plt.show()
```

```python
#salvar una imagen
plt.imsave('quetzalgris.png', Igray2)
```

## 4. Histogramas

```python
Igray2 = (255.*Igray2).astype(np.uint8)
```

```python
print(Igray2.min(), Igray2.max())
```

```python
Igray2.shape
```

Vectorizamos la imagen (la convertimos a un vector 1-dimensional:

```python
# vectorizar una imagen

Ivector = Igray2.ravel()
Ivector.shape
```

```python
minn = Ivector.min()
maxx = Ivector.max()
```

```python
h = np.histogram(Ivector, bins=maxx-minn+1)
```

```python
# histogramas

plt.figure()
plt.plot(np.arange(minn, 1+maxx), h[0])
#plt.hist(Igray2, bins=256)
plt.show()
```

```python
from scipy.ndimage import gaussian_filter
```

```python
Delta = np.zeros((101,101))
Delta[50,50] = 1.
```

```python
D = gaussian_filter(Delta, sigma=20.)
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.imshow(Delta, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(D, cmap='gray')
plt.show()
```

```python
# efecto blur sobre la imagen
J = gaussian_filter(Igray2, sigma=3.)
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.imshow(Igray2, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(J, cmap='gray')
plt.show()
```
