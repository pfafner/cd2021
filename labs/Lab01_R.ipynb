1. # Conjunto deport.dat  y PCA

```r
dataraw = read.table('deport.csv', sep=',', header=1)
head(dataraw)
```

```r
data = dataraw[,-9]
rownames(data) = dataraw[,9]
head(data)
```

```r
pairs(data, cex=0.5, pch=16)
rownames(data)[data[,2]>23]
colores = rep('blue', dim(data)[1])
colores[data[,2]>23]= 'red'
pairs(data, cex=0.8, pch=16, col=colores)
```

```r
library(GGally)
ggpairs(data, diag=list(continuous='densityDiag'), axisLabels='none')
```

```r
pca = princomp(data)
summary(pca)
loadings(pca)
```

```r
boxplot(data)
```

```r
pca = princomp(data, cor=T)
summary(pca)
loadings(pca)
```

```r
biplot(pca)
```

```r
Y = scale(data)
```

```r
pca2 = princomp(as.data.frame(Y))
summary(pca2)
loadings(pca2)
```

```r
biplot(pca2)
```
