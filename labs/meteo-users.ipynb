## 1. Leer datos

a) Leer el conjunto de datos 'meteo-users.csv'

Cambiar la fecha como columna índice.

```r
data = read.table('meteo-users.csv', sep=',', header=1)
head(data)
```

```r
# definimos la fecha commo índice del dataframe

rownames(data) = data[,1]
data = data[,2:8]
head(data)
```

## 2. Explorar datos

a) Explorar los datos

   Por ejemplo: bloxplots, pairplots, distribuciones, ...

```r
boxplot(data)
```

```r
boxplot(data[,1:6])
```

```r
pairs(data)
```

```r
library(GGally)
ggpairs(data, diag=list(continuous="densityDiag"), axisLabels="none")
```

b) Hacer una proyección de los datos sobre las primeras dos componentes.

# Estandarizar los datos


```r
data = read.table('meteo-users.csv', sep=',', header=1)
head(data)
```

```r
rownames(data) = data[,1]
data = data[,2:8]
head(data)
```

```r
mu = apply(data, 2, mean)
mu
```

```r
st = sqrt(apply(data, 2, var))
st
```

```r
# estandarizamos los datos
X = apply(data,2, function(x){(x - mean(x))/sqrt(var(x))})
head(X)
```

```r
# otra forma de estandarizar
Y = as.data.frame(scale(data))
head(Y)
```

```r
boxplot(Y)
```

```r
apply(Y, 2, mean)
```

```r
pca = princomp(Y)
summary(pca)
loadings(pca)
```

```r
biplot(pca)
```

```r
pairs(Y)
```

```r
ggpairs(Y, diag=list(continuous="densityDiag"), axisLabels="none")
```
